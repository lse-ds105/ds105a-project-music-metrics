## Function to get word count frequencies in string of lyrics

````python
def get_word_counts(words):
    word_list = words.split()
    word_counts = {}
    for word in word_list:
        if word in word_counts.keys():
            word_counts[word] +=1
        else:
            word_counts[word] = 1
    return word_counts
````
